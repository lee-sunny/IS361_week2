# Set Comprehensions

Set comprehensions are different from list comprehensions and dict comprehensions in that they are unordered collection of elements in which each element can only appear once.

## Syntax

The syntax for set comprehensions is almost identical to that of list comprehensions but it uses curly brackets instead of square brackets.




### Example 1: 
A set of squares of numbers 0-9

```python
nums = {n**2 for n in range(10)}
print(nums)
```

### Example 2:
A set of all of the prime numbers that are less than 100

```python
r= {x for x in range(2, 101) 
if not any(x % y == 0 for y in range(2, x))} 
print(r)
```
